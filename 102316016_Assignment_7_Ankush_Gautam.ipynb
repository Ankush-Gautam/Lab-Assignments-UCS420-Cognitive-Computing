```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Part 1
np.random.seed(42)
data = np.random.randint(1000, 5000, (12, 4))
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame(data, columns=categories, index=months)

#Part 2
print(df.head())
print(df.describe())
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100

random_number = np.random.randint(1, 100)
if random_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothing'] *= 0.85

#Part 3
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(months, df[category], label=category)
plt.legend()
plt.title('Monthly Sales Trends')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.show()

```
